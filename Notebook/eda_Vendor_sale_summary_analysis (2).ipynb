##### What is the 95% confidence interval for profit margins of top performing and low performing vendors

```python
df.columns
```

```python
top_thresold=df['total_sale_dollar'].quantile(0.75)
low_thresold=df['total_sale_dollar'].quantile(0.25)
```

```python
top_vendor=df[df['total_sale_dollar']>=top_thresold]['margin'].dropna()
low_vendor=df[df['total_sale_dollar']<=low_thresold]['margin'].dropna()
```

```python
top_vendor
```

```python
from scipy import stats
import numpy as np

def confidence_interval(data, confidence=0.95):
    mean_val = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(len(data))
    t_critical = stats.t.ppf((1 + confidence) / 2, df=len(data) - 1)
    margin_of_error = t_critical * std_err
    return mean_val, mean_val - margin_of_error, mean_val + margin_of_error
```

```python
top_mean,top_lower,top_upper=confidence_interval(top_vendor)
low_mean,low_lower,low_upper=confidence_interval(low_vendor)
print(f"Top Vendor 95% CI: ({top_lower:.2f},{top_upper:.2f}),Mean: {top_mean:.2f}")
print(f"Low Vendor 95% CI: ({low_lower:.2f},{low_upper:.2f}),Mean: {low_mean:.2f}")

plt.figure(figsize=(12,6))

# Top Vendor plot
sns.histplot(top_vendor,kde=True,color='blue',bins=30,alpha=0.5,label="Top Vendors")
plt.axvline(top_lower,color='blue',linestyle="-",label=f"top Lower:{top_lower:.2f}")
plt.axvline(top_upper,color='blue',linestyle="-",label=f"top Upper:{top_upper:.2f}")
plt.axvline(top_mean,color='blue',linestyle="-",label=f"top Mean:{top_mean:.2f}")

# Low Vendor plot
sns.histplot(low_vendor,kde=True,color='red',bins=30,alpha=0.5,label="low Vendors")
plt.axvline(low_lower,color='red',linestyle="-",label=f"low Lower:{low_lower:.2f}")
plt.axvline(low_upper,color='red',linestyle="-",label=f"low Upper:{low_upper:.2f}")
plt.axvline(low_mean,color='red',linestyle="-",label=f"low Mean:{low_mean:.2f}")

#finalize plot
plt.title("confindence Interval Comparision : Top vs Low Vendors (Profit:Margin)")
plt.xlabel("profit Margin (%)")
plt.ylabel("frequecy")
plt.legend()
plt.grid(True)
plt.show()
```

##### Interpretation

Low Vendors have a higher mean profit margin (≈ 42%) than top vendors (≈ 31%).

Top Vendors show tighter confidence intervals, meaning their profit margins are more consistent and predictable.

Low Vendors’ margins vary more, but they’re generally higher.

##### This suggests:

Top Vendors = stable, reliable performance (possibly due to higher volume, discounts, or strong contracts).

Low Vendors = higher profit per sale, but possibly lower volume or higher volatility.

##### Hypothesis Testing of Profit Margins Between Top and Low Vendors

```python
import numpy as np
import pandas as pd
from scipy import stats
import math

# assume top_vendor and low_vendor are 1-D numpy arrays or pandas Series
# top_vendor = np.array(...)
# low_vendor = np.array(...)

alpha = 0.05

# --- 1. Descriptive stats
def desc(x):
    return {
        "n": len(x),
        "mean": np.mean(x),
        "std": np.std(x, ddof=1),
        "median": np.median(x)
    }

print("Top vendors:", desc(top_vendor))
print("Low vendors:", desc(low_vendor))

# --- 2. Normality tests (Shapiro-Wilk)
sh_top = stats.shapiro(top_vendor)
sh_low = stats.shapiro(low_vendor)
print("\nShapiro Top: stat={:.4f}, p={:.4f}".format(sh_top.statistic, sh_top.pvalue))
print("Shapiro Low : stat={:.4f}, p={:.4f}".format(sh_low.statistic, sh_low.pvalue))

# --- 3. Variance equality test (Levene)
levene_stat, levene_p = stats.levene(top_vendor, low_vendor)
print("\nLevene: stat={:.4f}, p={:.4f}".format(levene_stat, levene_p))

# --- 4. Choose test & run t-test (Welch by default is safer)
# If you want classic Student's t-test when levene_p > alpha, set equal_var=True
equal_var = True if levene_p > alpha else False
t_stat, p_val = stats.ttest_ind(top_vendor, low_vendor, equal_var=equal_var)
print("\nT-test (equal_var={}): t={:.4f}, p={:.4f}".format(equal_var, t_stat, p_val))

# --- 5. Nonparametric alternative (if non-normal)
u_stat, u_p = stats.mannwhitneyu(top_vendor, low_vendor, alternative='two-sided')
print("\nMann-Whitney U: U={:.4f}, p={:.4f}".format(u_stat, u_p))

# --- 6. Cohen's d (effect size for t-test)
def cohens_d(a, b):
    na, nb = len(a), len(b)
    sa, sb = np.var(a, ddof=1), np.var(b, ddof=1)
    pooled_sd = math.sqrt(((na-1)*sa + (nb-1)*sb) / (na+nb-2))
    return (np.mean(a) - np.mean(b)) / pooled_sd

# When variances hugely different, use Glass's delta (use sd of control) — here's Cohen's d:
try:
    d = cohens_d(top_vendor, low_vendor)
    print("\nCohen's d (top - low): {:.4f}".format(d))
except Exception as e:
    print("\nCohen's d could not be computed:", e)

# --- 7. 95% CI for difference in means (Welch's approach)
def welch_ci(a, b, confidence=0.95):
    na, nb = len(a), len(b)
    ma, mb = np.mean(a), np.mean(b)
    sa, sb = np.var(a, ddof=1), np.var(b, ddof=1)
    se = math.sqrt(sa/na + sb/nb)
    # degrees of freedom (Welch-Satterthwaite)
    df = (sa/na + sb/nb)**2 / ((sa**2)/((na**2)*(na-1)) + (sb**2)/((nb**2)*(nb-1)))
    t_crit = stats.t.ppf((1+confidence)/2, df)
    diff = ma - mb
    lower = diff - t_crit*se
    upper = diff + t_crit*se
    return diff, lower, upper, df

diff, lower, upper, df = welch_ci(top_vendor, low_vendor)
print("\nDifference in means (top - low): {:.4f}".format(diff))
print("95% CI for difference: ({:.4f}, {:.4f}), df~{:.2f}".format(lower, upper, df))

# --- 8. Decision
if p_val < alpha:
    print("\nResult: reject H0 (statistically significant difference at alpha={})".format(alpha))
else:
    print("\nResult: fail to reject H0 (no statistically significant difference at alpha={})".format(alpha))

# If you used Mann-Whitney as main test, use u_p instead of p_val.

# --- Optional: print mean & CI for each group using t-distribution
def mean_ci(x, confidence=0.95):
    n = len(x)
    m = np.mean(x)
    se = np.std(x, ddof=1) / np.sqrt(n)
    t_crit = stats.t.ppf((1+confidence)/2, df=n-1)
    return m, m - t_crit*se, m + t_crit*se

print("\nTop mean CI:", mean_ci(top_vendor))
print("Low mean CI:", mean_ci(low_vendor))
```
